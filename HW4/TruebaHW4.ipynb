```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#### Problem 2 ####
def f(x):
    return 3 * np.cos(5 * x) - 2*x**3 + x**2 - 4*x + 16

def fp(x): 
    return -15*np.sin(5 * x) - 6*x**2 + 2 * x - 4

xlist = np.linspace(0,4,100)
flist = f(xlist)
fplist = fp(xlist)

h = .04040404

diflist = []
diflist.append((-25*flist[0]+48*flist[1]-36*flist[2]+16*flist[3]-3*flist[4])/(12*h))
diflist.append((-3*flist[0]-10*flist[1]+18*flist[2]-6*flist[3]+1*flist[4])/(12*h))

for i in range(2, 98):
    diflist.append((1*flist[i-2]-8*flist[i-1]+0*flist[i+0]+8*flist[i+1]-1*flist[i+2])/(12*1.0*h**1))

diflist.append((-1*flist[95]+6*flist[96]-18*flist[97]+10*flist[98]+3*flist[99])/(12*h))
diflist.append((3*flist[95]-16*flist[96]+36*flist[97]-48*flist[98]+25*flist[99])/(12*h))

plt.title('Real Derivative vs Finite Difference')
plt.scatter(xlist, diflist, label = 'Finite Difference', s = 6, color = 'orange')
plt.plot(xlist, fplist, label = 'Real Derivative')
plt.legend()

plt.show()

abserror = abs(diflist - fplist)

plt.semilogy(xlist, abserror)
plt.title('Absolute Error')
plt.show()
```
